```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load measurement data from file
npzfile = np.load('GPS_meas.npz', allow_pickle=True)
noisy_measurement = npzfile['measurements']
t_measurement = npzfile['t_measurements']
```

```python
plt.figure()
plt.suptitle('GPS measurements - POS XYZ')
plt.subplot(3, 1, 1)
plt.plot(noisy_measurement[:,0], 'r.-', label='Noisy measurement')
plt.subplot(3, 1, 2)
plt.plot(noisy_measurement[:,1], 'g.-', label='Noisy measurement')
plt.subplot(3, 1, 3)
plt.plot(noisy_measurement[:,2], 'b.-', label='Noisy measurement')
```
